**Data Cleaning**

//Create a new dataset by removing rows with missing values from the original dataset.

```python
new_dataset = dataset.dropna()
```

//Fill missing values for numeric columns ('float64' and 'int64' data types) with the median value

```python
numeric_clmns = new_dataset.select_dtypes(include=['float64', 'int64']).columns
```

```python
new_dataset[numeric_clmns] = new_dataset[numeric_clmns].fillna(new_dataset[numeric_clmns].median())
```

//Fill missing values for categorical columns ('object' data type) with the mode

```python
categorical_clmns = new_dataset.select_dtypes(include=['object']).columns
```

```python
new_dataset[categorical_clmns] = new_dataset[categorical_clmns].fillna(new_dataset[categorical_clmns].mode().iloc[0])
```

//Clean and convert the 'monthly_rent' column to 'int' for analysis

```python
new_dataset['monthly_rent'] = pd.to_numeric(dataset['monthly_rent'].astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')
new_dataset = new_dataset.dropna(subset=['monthly_rent'])
new_dataset['monthly_rent'] = new_dataset['monthly_rent'].astype(int)
```

//Clean and convert the 'size' column to 'int' for analysis

```python
new_dataset['size'] = new_dataset['size'].astype(str).str.extract('(\d+)').astype(int)
```

```python
new_dataset['rooms'] = new_dataset['rooms'].astype(str).str.extract('(\d+)').astype(int)
```

```python
new_dataset['bathroom'] = new_dataset['bathroom'].astype(str).str.extract('(\d+)').astype(int)
```

```python
new_dataset['parking'] = new_dataset['parking'].astype(str).str.extract('(\d+)').astype(int)
```

```python
new_dataset['completion_year'] = new_dataset['completion_year'].astype(int)
```

### **Data Visualization**

```python
property_type_counts = new_dataset['property_type'].value_counts()
print("\nProperty Type Distribution:")
print(property_type_counts)
```

```python
plt.figure(figsize=(15, 9))
ax = sns.barplot(x=property_type_counts.index, y=property_type_counts.values, color='green')
plt.xlabel("Property Type")
plt.ylabel("Count")
plt.title("Distribution of Property Types")
plt.xticks(rotation=90, ha="right")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(bottom=False)
plt.ylim(0, 3000)
plt.tight_layout()
for p in ax.patches:
      ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 8),
                textcoords='offset points')
plt.show()
```

```python
region_counts = new_dataset['region'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), textprops={'fontsize': 14})
plt.title('Distribution of Apartments by Region', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
properties = ['rooms', 'bathroom', 'parking']
colors = ['orange', 'skyblue', 'darkgreen']
labels = ['Rooms', 'Bathrooms', 'Parking']

x = 0
width = 0.25
for i, prop in enumerate(properties):
    plt.bar([p + width * i for p in range(len(new_dataset['property_type'].unique()))],
            new_dataset.groupby('property_type')[prop].mean(),
            width=width, color=colors[i], label=labels[i])

plt.xlabel('Property Type')
plt.ylabel('Average Count')
plt.title('Average Number of Rooms, Bathrooms, and Parking Spaces per Property Type')
plt.xticks([p + width for p in range(len(new_dataset['property_type'].unique()))], new_dataset['property_type'].unique(), rotation=90, ha='right')
plt.tick_params(bottom=False)
plt.ylim(0, 5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend()
plt.tight_layout()
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

```python
def plot_top_property_types_by_year(df, years):
    """Plots the top 3 property types for specified years."""
    for year in years:

        year_data = df[df['completion_year'] == year]

        if year_data.empty:
            print(f"No data found for the year {year}")
            continue


        top_types = year_data['property_type'].value_counts().nlargest(3)

        plt.figure(figsize=(8, 6))
        plt.bar(top_types.index, top_types.values, color='peachpuff')
        plt.xlabel("Property Type")
        plt.ylabel("Count")
        plt.title(f"Top 3 Property Types in {year}")
        plt.xticks(rotation=360, ha='center')
        plt.tick_params(bottom=False)
        plt.ylim(0, 500)
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['right'].set_visible(False)

        ax = plt.gca()
        for p in ax.patches:
            ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                textcoords='offset points')
        plt.tight_layout()
        plt.show()


plot_top_property_types_by_year(new_dataset, [2010, 2020])
```
